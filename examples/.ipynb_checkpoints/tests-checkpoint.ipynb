In this tutorial we are going to use Kinetic Monte Carlo method to study exciton dynmaics in Perylene-Red (PR) crystal.

First, we import all necessary libraries

```python
import numpy as np 
from core.tools import *
import importlib
from core.processes import couplings_neighbours, hopping_rate, add_exciton_random
from time import time
import ase
from ase.build import separate
from ase.io import read
from ase.visualize import view    
import math
from time import time
from warnings import warn
```

Here, we need o specify some input detailes to process the structure of the system we are going to study. Some of the parameteres related directly to the KMC process will be specified later.

Some properties i.e. excitation energy, transition dipole moments (TDM) should be calculated by ab-initio methods which we skip this part for this tutorial for simplicity. You can use any electronic structure program (Gausian, QChem, Orca, ...), but if you intend to stay completely in the python environment, we recommend using PySCF library. Very simple example using PySCF is also given in scf_calc.py file.


We use <b>Atom</b> object in <i>ASE</i> environment to specify each site (molecule). After uploading a structure file, you need to separate this Atom object into separate molecules. You can use <i>ase.build.separate(system)</i> for this purpose, but consider it is very time consuming especially if you have a big system. Instead, you can write your own function to separate them according to the atomic orders. This is exactly how we did it in this example: we called <i>separate_molecules_PBI()</i> function to do this. Note that this function can be used only for this system, create you own for your system.

```python
#============================================
""" USER INPUT """
#============================================

NUM_CPU = 8  #number CPUs for KMC calculation
FILENAME = 'PR_clean.pdb'  #input geometry file, *supported formats: vasp, pdb, cif, xyz. NOTE: data should include lattice information!
SUPERCELL = [10,10,10] # whether you want to create a supercell
CUTOFF = 20 # cut-off for interaction in A
mu0 = np.array([2.5634,     -0.0297,     -0.0109])   # Transition dipole moment in a.u.
J_overlap = 1.22      # spectral overlap, can be taken experiment
a0 = 0.529177 # Bohr radius, used to convert positions to a.u. #don't remove! 

#============================================
#============================================
""" Some post-processing procedure... """

for_cell = ase.io.read(FILENAME)  # getting molecule and cell information
print("Initializing calculation: we are using {} CPUs in total. \n".format(NUM_CPU))
s = for_cell.repeat(SUPERCELL)
print(SUPERCELL,"supercell created.")
start = time()
mol = separate_molecules_PBI(for_cell, SUPERCELL)            #   mol = ase.build.separate(s)   # separate Atoms into distinct molecules
print("System contains {} molecules.\n".format(len(mol)))
Vol = s.get_volume()*1e-12 # converting micrometr cube, match with experimental units
print("Volume: {:.3f} nm^3".format(Vol*1e9))
print("Lattice parameters [Å]:\n",for_cell.get_cell()[0], "\n",for_cell.get_cell()[1],"\n",for_cell.get_cell()[2],"\n")
t = time() - start
print("Time for molecule separation: {} min {:.2f} s".format(int(t/60), t-int(t/60)*60))
```

We created a list <i>mol</i> consisting individual molecules that we separated. Molecules could be oriented randomly which is different than the raw TDM data, because TDM calculated for monomer and we just specify this number here. You should be aware of that TDM can be oriented differently, e.g. along long molecular axis or short moleculear axis. Most of the organic molecules it is oriented along long molecular axis like our case. Here we use <i>add_tdm(mu0,mol)</i> function to rotate raw TDM to the long molecular axis and assign them as a feature with <i>tdm</i>. After executing this function, each molecule has a feature that you can have an access by <i>mol[0].tdm</i>

Now, we are ready to calculate electronic coupling between each pair in the system. We use <i>couplings_neighbours(molecules=mol, cutoff=20)</i> function to calculate it using Dipole-dipole approximation. Of course you are free to choose any other method, but consider it might be expensive. Alternatively, we provide another function called <i>TDENS_lowdin()</i> to calculate couplings by much more accurate Lowdin charge method.

Both functions return two values: neighbor list and coupling matrix. Neighbor list is the dictionary object with the neighbour molecules' indices for each given molecule. Coupling matrix is numpy.array object with float32. Since we usually have to deal with huge system with thousands of molecules, using float64 can be heavy for memory. This is very memory intensive especially when you try to run KMC calculation parallel. To reduce memory load, we need to convert np.array object to the sparse array by using Compressed Sparse Row method in <i>Scipy</i> library. Remember, when you introduce a cutoff, most of the values in coupling matrix will be zero, but it takes the same memory as float numbers, so in the sparse array we save only non-zero values, then it reduce RAM usage couple of hundred times at least. Also, make sure you delete the dense np.array object after conversion to the sparce array to free up RAM!

```python
from scipy.sparse import csr_matrix
import gc

mol = add_tdm(mu0, mol) # add rotated TDM to each molecule in the system

neighbors, V = couplings_neighbours(molecules=mol, cutoff=20)

V_csr = csr_matrix(V, dtype=V.dtype)      # keeps only the non–zero entries
print("dense aray :", V.shape, "   bytes:", V.nbytes / 1e6, "MB")
print("sparse array   :", V_csr.nnz,       "   bytes:", V_csr.data.nbytes / 1e6, "MB")

del V                                  # drop the dense array
gc.collect()                           # force garbage-collection cycle, free up the RAM

print("FINISH!")
```

Now we are almost ready to KMC calculation. So far, we have coupling matrix which will be used to calculate hopping rates by Forster rate theory. We also need decay and annihilation rate. For decay rate you can use Einstein radiative decay, but for latter you can choose a single value for all excitons. If you want instanteneous annihilation, then specify very large value, otherwise choose it wisely. (For organic molecules like pentacene, and some other perylene derivatives annihilation rate is in range of $10^{13} - 10^{15} \ s^{-1}$)

```python
from scipy.sparse import csr_matrix
from scipy.spatial import cKDTree
from tqdm import tqdm
from numba import jit


def couplings_neighbours_fast(molecules, cutoff=20):

    start = time()

    coords = np.array([mol.get_center_of_mass() for mol in molecules])
    kdtree = cKDTree(coords)

    # neighbours_idx[i] will be a list of j-indices within the cutoff,
    # including i itself (distance < 1e-3); query_ball_point is very fast.
    neighbours_idx = kdtree.query_ball_point(coords, cutoff)
    
    a0 = 0.529177   
    num_sites = len(molecules)
    rows, cols, data = [], [], []  
    #V = V.astype(np.float32) # reduce the size of the array. If high accuracy required, pls. use float64, but consider it is memory intensive!!!
    neighbors = {}
    for i in tqdm(range(0, num_sites)):
        nb = []
        center = molecules[i].get_center_of_mass()
        for j in neighbours_idx[i]:    # only j within cutoff
            if j <= i:        # skip lower triangle & diagonal
                continue
            target = molecules[j].get_center_of_mass()
            dist = np.linalg.norm(target-center)
            # print(i,j,np.linalg.norm(target-center)) # for debugging
            if dist < 1e-3:
                cc = 0
            elif dist <= cutoff:
                nb.append(j)
                #dipole approx. calculations
                D_i = molecules[i].tdm
                D_j = molecules[j].tdm
                k = ( np.dot(unit_vector(D_i), unit_vector(D_j)) 
                      - 3*np.dot(unit_vector((target-center)), unit_vector(D_i))*np.dot(unit_vector((target-center)), unit_vector(D_j)) )
                cc = k*np.linalg.norm(D_i)*np.linalg.norm(D_j)/((dist/a0)**3)
                rows += [i, j]               # symmetric entries
                cols += [j, i]
                data += [cc, cc]
            else:
                cc = 0     
        neighbors[i] = nb
    V_csr = csr_matrix((data, (rows, cols)),
                       shape=(num_sites, num_sites),
                       dtype=np.float32)
    t = time() - start
    print("Coupling calculation: done.")
    print("Neighbour list: done.")
    print("User time: {} min {:.2f} s".format(int(t/60), t-int(t/60)))
    return neighbors, V_csr 
```

```python

mol = add_tdm(mu0, mol) # add rotated TDM to each molecule in the system

neighbors, V = couplings_neighbours_fast(molecules=mol, cutoff=20)

#V_csr1 = csr_matrix(V, dtype=V.dtype)      # keeps only the non–zero entries
#print("dense aray :", V.shape, "   bytes:", V.nbytes / 1e6, "MB")
print("sparse array   :", V.nnz,       "   bytes:", V.data.nbytes / 1e6, "MB")

```

```python
"""diffusion constanc calculation """

import numpy as np
from scipy import stats      # only for linear fit; stdlib alternative shown below

def analyse_diffusion(all_runs, coords, dimension=3,
                      n_msd_bins=200, fit_fraction=(0.3, 0.6)):
    """
    Parameters
    ----------
    all_runs : list[ (times, exciton_records) ]
        Output of run_kmc_parallel: each element has
        times             (length T_i array)
        exciton_records   (length T_i list of list[int])
    coords : ndarray (N_sites, 3)
        xyz positions of every lattice site (same units you want for MSD).
    dimension : {1,2,3}
        Dimensionality 'd' for Einstein relation MSD = 2d D t
    n_msd_bins : int
        Number of bins for the common time grid when averaging MSD.
    fit_fraction : tuple(low, high)
        Fraction of the (0, t_max) time range to fit for the slope,
        e.g. (0.5,0.9) means use the last 50‑90 % for linear fit.

    Returns
    -------
    D   : diffusion coefficient   (units of coords² / time)
    tau : mean exciton lifetime   (time units)
    L   : diffusion length = sqrt(2 d D tau)  (length units)
    t_grid, msd_avg : arrays for the averaged MSD curve
    """

    # 1) Common time grid
    t_max = max(times[-1] for times, _ in all_runs)
    t_grid = np.linspace(0.0, t_max, n_msd_bins)

    msd_accum = np.zeros(n_msd_bins)
    msd_counts = np.zeros(n_msd_bins, dtype=int)
    lifetime_accum = []        # collect life times of individual excitons

    # --- helper to reconstruct one trajectory's exciton paths
    def reconstruct_times_xyz(times, excit_rec):
        # map each exciton to its xyz trajectory & lifetime
        first_sites = excit_rec[0]
        gid_map = {local: local for local in range(len(first_sites))}
        next_gid = len(first_sites)

        traj = {}          # gid -> [xyz0, xyz1, ...]
        ttraj = {}         # gid -> [t0, t1, ...]
        for local, gid in gid_map.items():
            traj[gid]  = [coords[first_sites[local]]]
            ttraj[gid] = [times[0]]

        for step in range(1, len(times)):
            s_time = times[step]
            sites  = excit_rec[step]
            new_map = {}
            for local_idx, site in enumerate(sites):
                if local_idx in gid_map:
                    gid = gid_map[local_idx]
                else:                       # new exciton appeared (unlikely)
                    gid = next_gid
                    next_gid += 1
                    traj[gid]  = []
                    ttraj[gid] = []
                new_map[local_idx] = gid
                traj[gid].append(coords[site])
                ttraj[gid].append(s_time)
            gid_map = new_map

        # compute lifetime list and MSD increments
        lifetime = [t_list[-1] - t_list[0] for t_list in ttraj.values()]
        return traj, ttraj, lifetime

    # --- pass through trajectories
    for times, excit_rec in all_runs:
        traj, ttraj, life = reconstruct_times_xyz(times, excit_rec)
        lifetime_accum.extend(life)

        # contribution to ensemble MSD
        for gid in traj:
            r0 = np.asarray(traj[gid][0])
            disp2 = [np.sum((np.asarray(r) - r0)**2) for r in traj[gid]]
            step_idx = np.searchsorted(t_grid, ttraj[gid], side='right') - 1
            step_idx[step_idx < 0] = 0
            for idx, d2 in zip(step_idx, disp2):
                msd_accum[idx] += d2
                msd_counts[idx] += 1

    # average
    mask = msd_counts > 0
    msd_avg = np.zeros_like(t_grid)
    msd_avg[mask] = msd_accum[mask] / msd_counts[mask]

    # 2) Diffusion coefficient from linear fit of MSD = 2 d D t
    low, high = fit_fraction
    fit_mask = (t_grid >= low * t_max) & (t_grid <= high * t_max) & mask
    if fit_mask.sum() < 2:
        raise RuntimeError("Not enough data points in chosen fit range.")
    slope, intercept, *_ = stats.linregress(t_grid[fit_mask], msd_avg[fit_mask])
    D = slope / (2 * dimension)

    # 3) Lifetime (mean)
    tau = np.mean(lifetime_accum)

    # 4) Diffusion length
    L = np.sqrt(2 * dimension * D * tau)

    return D, tau, L, t_grid, msd_avg
```

```python
D, tau, L, t_axis, msd_curve = analyse_diffusion(all_runs, coords)

print(f"Diffusion coefficient  D = {D*1e-5:.3e}  (nm²/ps)")
print(f"Mean lifetime          τ = {tau:.3e}  ns")
print(f"Diffusion length       L = {L:.3e}  (A")


import matplotlib.pyplot as plt
plt.plot(t_axis, msd_curve)
plt.xlabel('Time (ns)')
plt.ylabel('MSD (A²)')
plt.title('Averaged MSD')
plt.show()
```

```python
def reconstruct_runs_uid(times, records, coords):
    """
    records[t] = [(uid, site), (uid, site), ...]
    returns dict: uid -> (t_array, xyz_array)
    """
    traj = {}
    for step,(t, row) in enumerate(zip(times, records)):
        for uid, site in row:
            traj.setdefault(uid, ([],[]))
            traj[uid][0].append(t)
            traj[uid][1].append(coords[site])
    # convert to numpy
    for uid in traj:
        traj[uid] = (np.asarray(traj[uid][0]),
                     np.asarray(traj[uid][1]))
    return traj

reconstruct_runs_uid()
```

```python
def msd_single_run(times, exciton_records, coords):
    traj_dict = reconstruct_single_run(times, exciton_records, coords)

    # build a common time grid (here: the event times themselves)
    T = len(times)
    msd = np.zeros(T)
    counts = np.zeros(T, dtype=int)

    # for every exciton, compute its displacement r(t)‑r(0)²
    for t_arr, xyz_arr in traj_dict.values():
        r0 = xyz_arr[0]                # initial position for that exciton
        # map exciton's private times onto global step indices with searchsorted
        step_indices = np.searchsorted(times, t_arr)
        disp2 = np.sum((xyz_arr - r0)**2, axis=1)

        msd[step_indices]  += disp2
        counts[step_indices] += 1

    # avoid /0
    mask = counts > 0
    msd[mask] = msd[mask] / counts[mask]

    return np.asarray(times), msd

#a = msd_single_run(times_first, excitons_first, coords)
```

```python
def msd_many_runs(all_runs, coords, num_bins=200):
    # 1) common time grid (linear bins up to max end‑time)
    t_max = max(r[0][-1] for r in all_runs)
    t_grid = np.linspace(0.0, t_max, num_bins)

    msd_accum = np.zeros(num_bins)
    traj_count = np.zeros(num_bins, dtype=int)

    for times, exc_recs in all_runs:
        t_ev, msd_ev = msd_single_run(times, exc_recs, coords)
        # interpolate (step function) onto common grid
        idx = np.searchsorted(t_grid, t_ev, side='right')-1
        idx[idx < 0] = 0
        msd_accum[idx] += msd_ev
        traj_count[idx] += 1

    mask = traj_count > 0
    msd_accum[mask] = msd_accum[mask] / traj_count[mask]
    return t_grid, msd_accum
```

```python
import matplotlib.pyplot as plt

t_grid, msd_avg = msd_many_runs(all_runs, coords)

plt.plot(t_grid, msd_avg)      # time in ns, example
plt.xlabel('Time (ns)')
plt.ylabel(r'$\langle r^{2}(t) \rangle$ (nm$^{2}$)')
plt.title('Ensemble‑averaged MSD ({} trajectories)'.format(len(all_runs)))
plt.show()
```
